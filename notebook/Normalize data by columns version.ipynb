```python
#Load data
import pandas as pd # import Pandas library (and defining shorthand "pd") for reading and manipulating the data files
from matplotlib import pyplot as plt # import and define shorthand "plt" for library "pyplot" providing plotting functions
from IPython.display import display, HTML
import numpy as np   # import and define shorthand "np" for library "numpy" for advanced mathematical operations in python

#read in data from csv files 
#parse_dates function is used on Date-column to change them from string to date-object

x_train = pd.read_csv("train_data.csv", header= None) 
x_test= pd.read_csv("test_data.csv", header= None)
y = pd.read_csv("train_labels.csv", header= None) 
header= pd.read_csv("x_header.csv")
x_header= pd.DataFrame(data= header)
x_train.columns= x_header.ID
x_test.columns = x_header.ID
y.columns = ['Genre']
```

```python
display(HTML(x_train.head(5).to_html(max_rows=5)))
display(HTML(y.head(5).to_html(max_rows=5)))
```

```python
#append x_y_train
x_y_train = x_train.join(y)

columns = ['Rhythm_Mean', 'Rhythm_Median', 'Rhythm_Var', 'Rhythm_Kurtosis', 'Rhythm_Skewness','Rhythm_Min', 
           'Rhythm_Max', "Chroma_Mean", "Chroma_Std", "Chroma_Min", "Chroma_Max", "MFCC_Mean", "MFCC_Std", "MFCC_Min", "MFCC_Max"]

#Summarize data with mean
xysum = pd.DataFrame(columns = columns)
for label in columns:
    df = x_y_train.loc[:, x_y_train.columns.str.contains(label)].mean(axis=1)
    xysum[label] = df
xysum['Genre'] = x_y_train.Genre

#Groupby stat
print(xysum.groupby(xysum.Genre).describe())
```

```python
display(HTML(xysum.head(5).to_html(max_rows=5)))
```

```python
X = x_train
#see the frequency of all columns
for i in X.columns:
    print('Value count', i, 'frequency:', X[i].value_counts().head(5))

#From the data, we drop the columns with high frequencies of anomolies, which will affect the result in data analysis

X = X.drop(columns = ['Rhythm_Mean_24', 'Rhythm_Max_24','MFCC_Mean_1','MFCC_Mean_2','MFCC_Mean_3', 'MFCC_Mean_4', 'MFCC_Min_1'])
x_test = x_test.drop(columns = ['Rhythm_Mean_24', 'Rhythm_Max_24','MFCC_Mean_1','MFCC_Mean_2','MFCC_Mean_3', 'MFCC_Mean_4', 'MFCC_Min_1'])
```

```python
print(X.shape)
#data normalization
def normalize(X):
    x_norm_y = pd.DataFrame().reindex_like(X)
    x_norm_y.columns = X.columns
    x_norm_y.iloc[:,-1] = X.iloc[:,-1]
    for i in range(X.shape[1]-1):
        col_min = X.iloc[:,i].min()
        col_max = X.iloc[:,i].max()
        x_norm_y.iloc[:,i] = (X.iloc[:,i] - col_min)/(col_max-col_min)
    display(HTML(x_norm_y.head(5).to_html(max_rows=5)))
    return x_norm_y
    
X = normalize(X)
```

```python
#Create Validation set
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
y= np.ravel(y)
X_train, X_valid, y_train, y_valid = train_test_split(X, y,  test_size=0.33, random_state=42)

Logreg = LogisticRegression(penalty= 'l1',solver= 'saga',max_iter =500, multi_class='multinomial')

logreg = Logreg.fit(X_train, y_train)
print(Logreg.score(X_train, y_train))
y_train_pred =Logreg.predict(X_valid)
print(Logreg.score(X_valid, y_valid))
```

```python
#Cross validation
from sklearn.model_selection import cross_validate
from sklearn import metrics


crossValidation = cross_validate(Logreg, X_train, y_train, cv= 3)
print("Cross validation score", crossValidation['test_score'])
```

```python
x_test = normalize(x_test)
print(x_test.shape, type(x_test))

y_test = Logreg.predict(x_test)
```

```python
indx = x_test.index
result = pd.DataFrame(index = indx, columns = ["Sample_id","Sample_label"])
result["Sample_id"] = result.index+1
result["Sample_label"] = y_test
result.to_csv("solution.csv", index = False)
```
